```python

import numpy as np
import matplotlib.pyplot as plt

# Define function
def f(x):
    return 1.0 / (x - 2.0)

# Create x ranges approaching 2 from left and right
x_left = np.linspace(0.5, 1.999, 1000)
x_right = np.linspace(2.001, 3.5, 1000)

# Evaluate
y_left = f(x_left)
y_right = f(x_right)

plt.figure(figsize=(8,6))
plt.plot(x_left, y_left, label=r"$1/(x-2)$ (from left)", color='C0')
plt.plot(x_right, y_right, label=r"$1/(x-2)$ (from right)", color='C1')

# Vertical asymptote
plt.axvline(2, color='k', linestyle='--', linewidth=1)

# Annotations for infinities
plt.annotate(r"$\to -\infty$ as $x\to 2^{-}$",
             xy=(1.9, -10), xytext=(1.3, -15), arrowprops=dict(arrowstyle='->'))
plt.annotate(r"$\to +\infty$ as $x\to 2^{+}$",
             xy=(2.1, 10), xytext=(2.7, 15), arrowprops=dict(arrowstyle='->'))

plt.ylim(-50,50)
plt.xlim(0.5,3.5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('One-sided limits of f(x) = 1/(x-2) at x=2')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig('one_sided_limit_1_over_x_minus_2.png', dpi=150)
print('Saved one_sided_limit_1_over_x_minus_2.png')
```
